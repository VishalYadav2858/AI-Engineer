# creating dataframe for all the pages

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage= requests.get('https://www.ambitionbox.com/list-of-companies?page={}',headers=headers).text
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36"
}

final = pd.DataFrame()

for j in range(1, 91):

    # FIX 1: headers pass karo
    webpage = requests.get(
        f"https://www.ambitionbox.com/list-of-companies?page={j}",
        headers=headers
    ).text

    soup = BeautifulSoup(webpage, "lxml")

    # FIX 2: new class name
    company = soup.find_all(class_="companyCardWrapper")

    name = []
    rating = []
    reviews = []
    ctype = []
    hq = []

    for i in company:

        try:
            name.append(i.find("h2").text.strip())
        except:
            name.append(np.nan)

        try:
            # FIX 3: rating_text
            rating.append(i.find(class_="rating_text").text.strip())
        except:
            rating.append(np.nan)

        try:
            # FIX 4: new review class
            reviews.append(i.find(class_="companyCardWrapper__companyRatingCount").text.strip())
        except:
            reviews.append(np.nan)

        try:
            # FIX 5: split first interLinking element
            info = i.find_all(class_="companyCardWrapper__interLinking")[0].text.strip().split("|")
            ctype.append(info[0].strip())
        except:
            ctype.append(np.nan)

        try:
            # FIX 6: HQ from same split
            info = i.find_all(class_="companyCardWrapper__interLinking")[0].text.strip().split("|")
            hq.append(info[1].strip())
        except:
            hq.append(np.nan)

    df_1 = pd.DataFrame({
        "name": name,
        "rating": rating,
        "reviews": reviews,
        "company_type": ctype,
        "Head_Quarters": hq
    })

    final = pd.concat([final, df_1], ignore_index=True)

print(final.head())
print(final.shape)
```
